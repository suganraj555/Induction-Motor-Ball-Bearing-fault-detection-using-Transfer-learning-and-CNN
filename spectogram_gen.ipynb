```python
import scipy.io
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
from os import listdir
from os.path import isfile, join
import os
import matplotlib
matplotlib.use('Agg')
```

```python
def extract_name(file):
  c = 0
  for i in file:
    if(i.isalpha()):
      c+=1
    else:
      break
  return file[:c]
```

```python

def create_image(dir_path,label,path):
    try:
        mat = scipy.io.loadmat(dir_path+"/"+path,appendmat=False)
    except:
        print("Exception occured")
        return

    if("data" not in mat.keys() and "fs" not in mat.keys() ):
        print("image could not be generated")
        print("Cols are ",mat.keys())
        return

    mat_data = mat['data']
    fs = mat['fs']

    num_samples, num_channels = mat_data.shape

    size = int(num_samples/100)

    for i in range(0,num_samples,size):
      data=mat_data[i:i+size]

    # Create a figure
      fig = plt.figure(frameon = False,clear=True)
      fig.set_size_inches(12,8)
      ax1 = plt.Axes(fig,[0.,0.,1.,1.])
      ax1.set_axis_off()
      fig.add_axes(ax1)
      #fig, ax1 = plt.subplots(figsize=(12, 8))

      data_for_specgram = data[:, 0]
      spec, freqs, t, im = ax1.specgram(data_for_specgram, Fs=float(fs), NFFT=256, cmap='magma')
      # ax1.set_title('Spectrogram of Sensor Data')
      # ax1.set_xlabel('Time')
      # ax1.set_ylabel('Frequency')

    # Save the image
      idx = int(i/size)
      plt.savefig("/content/drive/MyDrive/imgs/"+label+"/"+path[:-4]+"_"+str(idx)+"_img.png")

  # Display the image (optional)
#    plt.show()


      plt.close(fig)

```

```python
def create_folder(name):
  path = "/content/drive/MyDrive/imgs/"+name
  try:
    os.mkdir(path)
  except:
    print("Error in creating path for "+name)
```

```python
path = "/content/drive/MyDrive/dataset"

files = [f for f in listdir(path) if isfile(join(path, f))]

s = []

print(len(files))

for file in files:
    label = extract_name(file)
    if(label not in s):
      s.append(label)
      create_folder(label)
    create_image(path,label,file)
```
